```python
#wczytujemy plik csv, data parser czyli ze plik zawiera daty#

data = pd.read_csv('BTC-USD.csv', date_parser = True)
```

```python
#wyswietlam dane z csv#

data.head()
```

```python
#definiuje które dane maja byc danymi treningowymi (z jakiego okresu czasu)#

data_training = data[data['Date']< '2020-07-07']
data_training
```

```python
#definiuje ktore dane maja byc danymi testowymi (z jakiego okresu czasu)#

data_test = data[data['Date']>= '2020-07-07']
data_test
```

```python
#wyrzucam z danych treningowych kolumne z datą i adj close#

data_training = data_training.drop(['Date', 'Adj Close'], axis = 1)
data_training.head()
```

```python
#dane treningowe wskoczyły nam na wartości miedzy 1 i 0#

scaler = MinMaxScaler()
data_training = scaler.fit_transform(data_training)
data_training
```

```python
#

X_train = []
y_train = []

for i in range(60, data_training.shape[0]):
    X_train.append(data_training[i-60:i])
    y_train.append(data_training[i, 0])
    
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train.shape, y_train.shape
```

```python
# lSTM
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dropout, Dense, LSTM
```

```python
regressor = Sequential()

regressor.add(LSTM(units = 60, activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1], 5)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 120, activation = 'relu'))
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 1))
```

```python
regressor.summary()
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressor.fit(X_train, y_train, epochs = 10, batch_size = 32)
```
